```python
'''
Remover colunas com muitos valores ausentes

 nas colunas 'matricula_docente_externo', 'observacao', 'id_turma_agrupadora', 'convenio' possuem muitos valores nulos que não
 seram utilizados nas analises, então pode-se exclui-lás
'''
colunas_para_remover = ['matricula_docente_externo', 'observacao', 'id_turma_agrupadora', 'convenio']
df1.drop(columns=colunas_para_remover, inplace=True)

# Verificar o resultado
display(df1.info())
display(df1.isnull().sum())
```

```python
""" Verificando valores únicos nas colunas com valores ausentes em df1
  Analisando as colunas que ainda possuem "buracos" e quais são seus indices
"""
buracos = ['siape', 'campus_turma', 'local', 'descricao_horario', 'total_solicitacoes', 'capacidade_aluno', 'data_consolidacao', 'qtd_aulas_lancadas']

for col in buracos:
    print(f"Coluna: {col}")
    display(df1[col].value_counts(dropna=False))
    print("-" * 30)
```

```python
# Tratando valores ausentes em df1

# Para colunas com valores numéricos que podem ser preenchidos com a mediana ou média
# 'total_solicitacoes', 'capacidade_aluno', 'qtd_aulas_lancadas'
df1['total_solicitacoes'].fillna(df1['total_solicitacoes'].median(), inplace=True)
df1['capacidade_aluno'].fillna(df1['capacidade_aluno'].median(), inplace=True)
df1['qtd_aulas_lancadas'].fillna(df1['qtd_aulas_lancadas'].median(), inplace=True)

# Para colunas com valores categóricos que podem ser preenchidos com a moda
# 'campus_turma', 'local', 'descricao_horario', 'situacao_turma', 'modalidade_participantes'
# Note: 'situacao_turma' and 'modalidade_participantes' had no missing values according to .isnull().sum() but are included here for completeness if needed later.
for col in ['campus_turma', 'local', 'descricao_horario']:
    if df1[col].isnull().any():
        most_frequent_value = df1[col].mode()[0]
        df1[col].fillna(most_frequent_value, inplace=True)

# 'siape' e 'data_consolidacao' - estas colunas podem requerer um tratamento mais específico dependendo da análise.
# Por enquanto, vamos preencher 'siape' com 0 (indicando docente externo ou sem informação) e 'data_consolidacao' com um valor placeholder ou deixá-lo como está se não for crucial para a análise.
# Para 'siape', preencher com 0 parece razoável se quisermos manter a linha e indicar a ausência do SIAPE.
df1['siape'].fillna(0, inplace=True)

# Para 'data_consolidacao', se a data for importante, podemos tentar imputar com a data mais frequente ou uma data indicativa de "não consolidado".
# Dado que é uma data/hora, a imputação com a moda pode não ser significativa. Vamos preenchê-la com uma string indicando "Não Consolidado" por enquanto.
df1['data_consolidacao'].fillna('Não Consolidado', inplace=True)


# Verificar se ainda existem valores ausentes
display(df1.isnull().sum())
display(df1.info())
```

```python
# Tratando valores ausentes restantes em df2

# Para colunas com valores numéricos
df2['total_solicitacoes'] = df2['total_solicitacoes'].fillna(df2['total_solicitacoes'].median())
df2['capacidade_aluno'] = df2['capacidade_aluno'].fillna(df2['capacidade_aluno'].median())
df2['qtd_aulas_lancadas'] = df2['qtd_aulas_lancadas'].fillna(df2['qtd_aulas_lancadas'].median())

# Para colunas com valores categóricos
for col in ['campus_turma', 'local', 'descricao_horario']:
    if df2[col].isnull().any():
        most_frequent_value = df2[col].mode()[0]
        df2[col] = df2[col].fillna(most_frequent_value)

# Coluna 'siape'
df2['siape'] = df2['siape'].fillna(0)

# Coluna 'data_consolidacao'
df2['data_consolidacao'] = df2['data_consolidacao'].fillna('Não Consolidado')

# Verificar se ainda existem valores ausentes em df2
print("\nValores ausentes em df2 após tratamento:")
display(df2.isnull().sum())
display(df2.info())
```

```python
# Investigating anomalies in 'ch_dedicada_periodo'
Q1_ch = df['ch_dedicada_periodo'].quantile(0.25)
Q3_ch = df['ch_dedicada_periodo'].quantile(0.75)
IQR_ch = Q3_ch - Q1_ch
outlier_threshold_ch = Q3_ch + 1.5 * IQR_ch

print("Linhas com valores atípicos em 'ch_dedicada_periodo':")
display(df[df['ch_dedicada_periodo'] > outlier_threshold_ch])

# Investigating anomalies in 'capacidade_aluno'
Q1_cap = df['capacidade_aluno'].quantile(0.25)
Q3_cap = df['capacidade_aluno'].quantile(0.75)
IQR_cap = Q3_cap - Q1_cap
outlier_threshold_cap = Q3_cap + 1.5 * IQR_cap

print("\nLinhas com valores atípicos em 'capacidade_aluno':")
display(df[df['capacidade_aluno'] > outlier_threshold_cap])

# Investigating anomalies in 'total_solicitacoes'
Q1_sol = df['total_solicitacoes'].quantile(0.25)
Q3_sol = df['total_solicitacoes'].quantile(0.75)
IQR_sol = Q3_sol - Q1_sol
outlier_threshold_sol = Q3_sol + 1.5 * IQR_sol

print("\nLinhas com valores atípicos em 'total_solicitacoes':")
display(df[df['total_solicitacoes'] > outlier_threshold_sol])
```

```python
columns_with_outliers = ['ch_dedicada_periodo', 'capacidade_aluno', 'total_solicitacoes']

print("Média das colunas com possíveis outliers:")
display(df[columns_with_outliers].mean())
```

# Dia 03: Data Mining e Clustering, extraindo novas informações.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Carregar o dataframe combinado
df_tudo = pd.read_csv("/content/df_tudo.csv")
df = df_tudo.copy() # Criar uma cópia para trabalhar e evitar modificar o dataframe original
```

 Escolhi 4 variáveis por serem numéricas e representarem características importantes das turmas (capacidade, procura e carga horária) e a modalidade que se mostrou relevante na AED.

```python
colunas_numericas = ['ch_dedicada_periodo', 'capacidade_aluno', 'total_solicitacoes']
colunas_categoricas = ['modalidade_participantes'] # Mantendo a modalidade que se mostrou relevante na AED


# Criar transformadores para escalonamento e codificação
transformador_numerico = StandardScaler()
transformador_categorico = OneHotEncoder(handle_unknown='ignore') # Ignorar categorias desconhecidas durante a transformação

# Combinar os transformadores usando ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', transformador_numerico, colunas_numericas),
        ('cat', transformador_categorico, colunas_categoricas)
    ],
    remainder='passthrough' # Manter outras colunas, se houver (não aplicável neste caso específico, mas boa prática)
)

# Criar um pipeline que aplica o pré-processamento
pipeline_preprocessamento = Pipeline(steps=[('preprocessor', preprocessor)])

# Aplicar o pipeline aos dados selecionados
df_processed = pipeline_preprocessamento.fit_transform(df[colunas_numericas + colunas_categoricas])

# Converter o resultado de volta para um DataFrame (opcional, mas útil para visualização e análise)
# Obter os nomes das colunas após o one-hot encoding
nomes_colunas_categoricas = preprocessor.named_transformers_['cat'].get_feature_names_out(colunas_categoricas)
nomes_colunas_processadas = colunas_numericas + list(nomes_colunas_categoricas)

df_processed = pd.DataFrame(df_processed, columns=nomes_colunas_processadas)


# Exibir as primeiras linhas do dataframe processado e suas informações
print("Primeiras 5 linhas do dataframe processado:")
display(df_processed)
```

```python
# @title capacidade_aluno vs total_solicitacoes

from matplotlib import pyplot as plt
df_processed.plot(kind='scatter', x='capacidade_aluno', y='total_solicitacoes', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

## 2. Determinar o Número Ótimo de Clusters (Método do Cotovelo)

Aplicar o método do cotovelo para determinar o número ideal de clusters para o K-means e visualizar o resultado com um gráfico. Justificar a escolha do número de clusters.

**Reasoning**:
Determine the optimal number of clusters for the processed data using the elbow method. This involves calculating the sum of squared distances for a range of cluster numbers and visualizing the results.

```python
# Lista para armazenar a soma dos quadrados das distâncias (WCSS) para diferentes números de clusters
wcss = []

# Definir o intervalo de números de clusters a serem testados
# Geralmente, testamos de 1 até um número razoável (por exemplo, 10 ou 15)
k_range = range(1, 11)

# Calcular o WCSS para cada número de clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init para evitar problemas de inicialização
    kmeans.fit(df_processed)
    wcss.append(kmeans.inertia_) # inertia_ é o WCSS

# Plotar o gráfico do método do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, marker='o')
plt.title('Método do Cotovelo para Determinar o Número Ótimo de Clusters')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('WCSS (Soma dos Quadrados das Distâncias)')
plt.xticks(k_range) # Garante que todos os valores de K testados apareçam no eixo X
plt.grid(True)
plt.show()

# Justificativa para a escolha do número de clusters será feita após a visualização do gráfico.
```

## 3. Aplicar o K-Means e Adicionar Rótulos de Cluster

Aplicar o algoritmo K-Means com o número ótimo de clusters determinado na etapa anterior e adicionar os rótulos dos clusters ao dataframe original para análise posterior.

**Reasoning**:
Apply the K-Means algorithm with the chosen number of clusters (K=4) to the processed data and add the resulting cluster labels back to the original dataframe for further analysis and interpretation of the clusters.

```python
# Aplicar o K-Means com o número ótimo de clusters (K=4)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(df_processed)

# Adicionar os rótulos dos clusters ao dataframe original
df['cluster_label'] = kmeans.labels_

# Exibir as primeiras linhas do dataframe com os rótulos dos clusters
print("Primeiras 5 linhas do dataframe com rótulos dos clusters:")
display(df.head())

# Exibir a contagem de turmas em cada cluster
print("\nContagem de turmas por cluster:")
display(df['cluster_label'].value_counts())
```

# Dia 04: Testes e Hipóteses


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Carregar o dataframe combinado
df_tudo = pd.read_csv("/content/df_tudo.csv")
df = df_tudo.copy() # Criar uma cópia para trabalhar e evitar modificar o dataframe original
```

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Hipótese 1: Diferença na capacidade média de alunos entre níveis de ensino (LATO SENSU, GRADUAÇÃO, STRICTO SENSU)
# Como estamos comparando as médias de mais de dois grupos independentes (LATO SENSU, GRADUAÇÃO, STRICTO SENSU) e a variável dependente ('capacidade_aluno') é numérica, o teste ANOVA é apropriado se as premissas forem atendidas. Vamos verificar a normalidade e a homogeneidade das variâncias, embora a ANOVA seja razoavelmente robusta a desvios moderados.

# Vamos primeiro verificar a normalidade visualmente com histogramas para cada grupo.
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df[df['nivel_ensino'] == 'LATO SENSU']['capacidade_aluno'], kde=True)
plt.title('Distribuição da Capacidade (LATO SENSU)')

plt.subplot(1, 3, 2)
sns.histplot(df[df['nivel_ensino'] == 'GRADUAÇÃO']['capacidade_aluno'], kde=True)
plt.title('Distribuição da Capacidade (GRADUAÇÃO)')

plt.subplot(1, 3, 3)
sns.histplot(df[df['nivel_ensino'] == 'STRICTO SENSU']['capacidade_aluno'], kde=True)
plt.title('Distribuição da Capacidade (STRICTO SENSU)')

plt.tight_layout()
plt.show()

# A partir dos histogramas, parece que a distribuição da capacidade de aluno não é perfeitamente normal em todos os grupos. No entanto, a ANOVA é razoavelmente robusta. Vamos prosseguir com a ANOVA.

lato_sensu_capacity = df[df['nivel_ensino'] == 'LATO SENSU']['capacidade_aluno']
graduacao_capacity = df[df['nivel_ensino'] == 'GRADUAÇÃO']['capacidade_aluno']
stricto_sensu_capacity = df[df['nivel_ensino'] == 'STRICTO SENSU']['capacidade_aluno']

# Realizar o teste ANOVA
f_statistic_capacity, p_value_capacity = stats.f_oneway(lato_sensu_capacity, graduacao_capacity, stricto_sensu_capacity)

print("Resultado do Teste ANOVA para Capacidade Média por Nível de Ensino:")
print(f"Estatística F: {f_statistic_capacity:.2f}")
print(f"Valor p: {p_value_capacity:.4f}")
```

**Conclusão do Teste ANOVA:**

Usando o teste tipo ANOVA dá  para ver que a capacidade média dos alunos mudava de acordo com o nível de ensino: Lato Sensu, Graduação e Stricto Sensu. O resultado mostrou um valor de p bem baixo (menor que 0,05), o que indica que as diferenças entre os grupos não são por acaso. Isso quer dizer que, em pelo menos dois desses níveis, a capacidade média dos alunos é realmente diferente. Aliás, olhando os histogramas, essa diferença já parecia bem clara.

```python
# Hipótese 2: Diferença no total médio de solicitações entre modalidades Presencial e A Distância
# Como estamos comparando as médias de dois grupos independentes (Presencial e A Distância) e a variável dependente ('total_solicitacoes') é numérica, podemos usar um teste t independente. Vamos verificar a normalidade e homogeneidade das variâncias.

# Verificar normalidade visualmente com histogramas
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df[df['modalidade_participantes'] == 'Presencial']['total_solicitacoes'], kde=True)
plt.title('Distribuição de Solicitações (Presencial)')

plt.subplot(1, 2, 2)
sns.histplot(df[df['modalidade_participantes'] == 'A Distância']['total_solicitacoes'], kde=True)
plt.title('Distribuição de Solicitações (A Distância)')

plt.tight_layout()
plt.show()

# As distribuições parecem assimétricas. Um teste t independente pode não ser o mais adequado.
# Podemos usar o teste de Mann-Whitney U, que é um teste não paramétrico para comparar as distribuições de dois grupos independentes.

presencial_solicitacoes = df[df['modalidade_participantes'] == 'Presencial']['total_solicitacoes']
distancia_solicitacoes = df[df['modalidade_participantes'] == 'A Distância']['total_solicitacoes']

# Realizar o teste de Mann-Whitney U
u_statistic_solicitacoes, p_value_solicitacoes = stats.mannwhitneyu(presencial_solicitacoes, distancia_solicitacoes)

print("Resultado do Teste de Mann-Whitney U para Total de Solicitações por Modalidade:")
print(f"Estatística U: {u_statistic_solicitacoes:.2f}")
print(f"Valor p: {p_value_solicitacoes:.4f}")
```

**Conclusão do Teste de Mann-Whitney U:**

Como a distribuição do total de solicitações não parecia seguir um padrão normal, utilizamos o teste não paramétrico de Mann-Whitney U para comparar as modalidades Presencial e A Distância. O resultado apresentou um valor de p muito baixo (menor que 0,05), mostrando que existe uma diferença estatisticamente significativa entre as duas modalidades. Em outras palavras, o padrão de solicitações nas turmas a distância é realmente diferente daquele observado nas turmas presenciais — algo que já havia sido sugerido pela análise visual dos dados.

```python
# Hipótese 3: Diferença na distribuição da carga horária dedicada por nível de ensino
# Como estamos comparando as distribuições de uma variável numérica ('ch_dedicada_periodo') entre múltiplos grupos independentes (os diferentes níveis de ensino) e a normalidade não é clara, o teste de Kruskal-Wallis é uma escolha robusta e não paramétrica.

# Vamos filtrar os dados de carga horária para cada nível de ensino presente no dataframe.
formacao_complementar_ch = df[df['nivel_ensino'] == 'FORMAÇÃO COMPLEMENTAR']['ch_dedicada_periodo']
graduacao_ch = df[df['nivel_ensino'] == 'GRADUAÇÃO']['ch_dedicada_periodo']
lato_sensu_ch = df[df['nivel_ensino'] == 'LATO SENSU']['ch_dedicada_periodo']
stricto_sensu_ch = df[df['nivel_ensino'] == 'STRICTO SENSU']['ch_dedicada_periodo']
tecnico_ch = df[df['nivel_ensino'] == 'TÉCNICO']['ch_dedicada_periodo']

# Realizar o teste Kruskal-Wallis
h_statistic_ch, p_value_ch = stats.kruskal(formacao_complementar_ch, graduacao_ch, lato_sensu_ch, stricto_sensu_ch, tecnico_ch)

print("Resultado do Teste Kruskal-Wallis para Distribuição da Carga Horária por Nível de Ensino:")
print(f"Estatística H: {h_statistic_ch:.2f}")
print(f"Valor p: {p_value_ch:.4f}")
```

**Conclusão do Teste Kruskal-Wallis:**

Aplicamos o teste de Kruskal-Wallis para verificar se a carga horária dedicada por período varia entre os diferentes níveis de ensino. O resultado trouxe um valor de p extremamente baixo (menor que 0,05), o que nos permitiu rejeitar a hipótese nula.
